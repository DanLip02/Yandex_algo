```python
def main():
    """
    Пример ввода и вывода числа n, где -10^9 < n < 10^9:
    n = int(input())
    print(n)
    """
    s = input()
    result = []
    i = 0
    while i < len(s):
        # Если через 2 символа стоит '#' — это двузначное число (10-26)
        if i + 2 < len(s) and s[i + 2] == '#':
            num = int(s[i:i+2])
            result.append(chr(ord('a') + num - 1))
            i += 3
        else:
            num = int(s[i])
            result.append(chr(ord('a') + num - 1))
            i += 1
    
    print(''.join(result))

if __name__ == '__main__':
    main()
```
